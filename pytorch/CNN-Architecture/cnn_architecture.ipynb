```python
from torchinfo import summary
from LeNet import LeNet
from AlexNet import AlexNet
from VGGNet import VGGNet
```

### LeNet

```python
leNet=LeNet()
summary(leNet, input_size=(1, 1, 32, 32), col_names= ("input_size", "kernel_size", "output_size", "num_params", "mult_adds"), verbose=2)
```

### AlexNet

```python
alex_net_model=AlexNet()
summary(alex_net_model, input_size=(1, 3, 227, 227), col_names= ("input_size", "kernel_size", "output_size", "num_params", "mult_adds"), verbose=2)
```

### VGG

```python
model = VGGNet(in_channels=3, num_classes=1000)
summary(model, input_size=(1,3,224,224), col_names= ("input_size", "kernel_size", "output_size", "num_params", "mult_adds"), verbose=2)

x=torch.randn(1,3,224,224)
print(model(x).shape)
```
